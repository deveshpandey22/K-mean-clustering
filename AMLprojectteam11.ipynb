What is Meant by the K-Means Clustering Algorithm?

K-Means clustering is an unsupervised learning algorithm. 
There is no labeled data for this clustering, unlike in supervised learning. 



K-Means performs the division of objects into clusters that share similarities and are dissimilar to the objects belonging to another cluster. 

The term ‘K’ is a number. You need to tell the system how many clusters you need to create. For example, K = 2 refers to two clusters. There is a way of finding out what is the best or optimum value of K for a given data. 

For a better understanding of k-means, let's take an example from cricket. 



Imagine you received data on a lot of cricket players from all over the world, which gives information on the runs scored by the player and the wickets taken by them in the last ten matches. 



Based on this information, we need to group the data into two clusters, namely batsman and bowlers.

**The first step** in k-means clustering is the allocation of two centroids randomly (as K=2). Two points are assigned as centroids. Note that the points can be anywhere, as they are random points. They are called centroids, but initially, they are not the central point of a given data set
centroids1.jfif

The next step is to determine the distance between each of the randomly assigned centroids' data points. For every point, the distance is measured from both the centroids, and whichever distance is less, that point is assigned to that centroid. 
centroids2.jfif

The next step is to determine the actual centroid for these two clusters. The original randomly allocated centroid is to be repositioned to the actual centroid of the clusters.

centroids3.jfif

This process of calculating the distance and repositioning the centroid continues until we obtain our final cluster. Then the centroid repositioning stops.centroids4.jfif

How Does K-Means Clustering Work?
The flowchart below shows how k-means clustering works:FLowchart.jfif

The goal of the K-Means algorithm is to find clusters in the given input data. There are a couple of ways to accomplish this. We can use the trial and error method by specifying the value of K (e.g., 3,4, 5). As we progress, we keep changing the value until we get the best clusters. 

Another method is to use the **Elbow technique** to determine the value of K.
Step 1:
The Elbow method is the best way to find the number of clusters. The elbow method constitutes running  K-Means clustering on the dataset.

Next, we use within-sum-of-squares as a measure to find the optimum number of clusters that can be formed for a given data set. Within the sum of squares (WSS) is defined as the sum of the squared distance between each member of the cluster and its centroid.Elbow1.jfif
The WSS is measured for each value of K. The value of K, which has the least amount of WSS, is taken as the optimum value.

Now, we draw a curve between WSS and the number of clusters.

We randomly pick K (centroids). We name them c1,c2,..... ck, and we can say that centoidc1.jfif

We assign each data point to its nearest center, which is accomplished by calculating the euclidean distance.

We identify the actual centroid by taking the average of all the points assigned to that cluster.

**5. Model Building**

```python
# k-means with some arbitrary k

kmeans = KMeans(n_clusters=4, n_init=10 ,max_iter=50)
kmeans.fit(var_list)
kmeans.labels_
```

n_clusters: The number of clusters to form. In this case, the value is set to 4, meaning that the algorithm will group the data into 4 clusters.

n_init: The number of times the KMeans algorithm will be run with different centroid seeds. The default value for this parameter is 10, meaning that the algorithm will be run 10 times with different initial centroids to find the best clustering solution.

max_iter: The maximum number of iterations for each run of the KMeans algorithm. The default value for this parameter is 300, but in this case, it has been set to 50, meaning that the algorithm will stop clustering if it reaches 50 iterations before converging.

The data points have been grouped into 4 clusters, represented by the values 0, 1, 2, and 3

```python
ssd = []

for num_clusters in range(1,11):
    kmeans = KMeans(n_clusters=num_clusters, n_init=10, max_iter=50)
    kmeans.fit(var_list)
    
    ssd.append(kmeans.inertia_)
```

The elbow method involves plotting the Sum of Squared Distances (SSD) for each cluster solution against the number of clusters used. The "elbow point" is the point on the curve where the rate of decrease in SSD begins to level off. This point is considered to be the optimal number of clusters for the data set.

```python
plt.figure(figsize=(12,6))

plt.plot(range(1,11), ssd, linewidth=2, color="red", marker ="8")
plt.title("Elbow Curve\n=================================================================", fontsize=20, color="green")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("SSD")

plt.show()
```

Here we can see that from K value **5** the rate of decrease in SSD begins to level off. So we are going to make 5 cluster for the dataset

```python
kmeans = KMeans(n_clusters=5, n_init=10, max_iter=50)
kmeans.fit(var_list)
```

```python
kmeans.labels_
```

The data points have been grouped into 5 clusters, represented by the values 0, 1, 2, 3 and 4

```python
df["Label"] = kmeans.labels_
df.head()
```

```python
# Select the variables to use for clustering
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values

# Scale the variables if necessary
# You can use the StandardScaler from sklearn.preprocessing to do this

# Choose the number of clusters to create
k = 5

# Apply the k-means clustering algorithm
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Analyze the results
df['Cluster'] = y_kmeans
print(df.head())

# Visualize the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
plt.show()
```
